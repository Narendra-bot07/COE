```python
from scipy.stats import bernoulli
```

```python
p=0.6
data = bernoulli.rvs(p,size=1000)
```

```python
plt.bar([0,1],[sum(data==0),sum(data==1)],color=['r','g'],tick_label=['failure[0]','sucess(1)'])
plt.xlabel('outcome')
plt.ylabel('frequency')
plt.title('bernoulli distribution(0.6)')
plt.show()
```

```python
from scipy.stats import binom
```

```python
n=10
p=0.3
x = np.arange(0,n+1)
pmf = binom.pmf(x,n,p)
plt.bar(x,pmf,color=['blue'],alpha=0.7)
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('binomial distribution')
plt.show()
```

```python
from scipy.stats import poisson
```

```python
lam = 3
x = np.arange(0,10)
pmf = poisson.pmf(x,lam)
```

```python
plt.bar(x,pmf,color='purple',alpha=0.4)
plt.xlabel('Number of calls per minute')
plt.ylabel('probability')
plt.title('Poisson distribution')
```

```python
from scipy.stats import norm
```

```python
mu = 70
sigma = 10
x = np.linspace(mu-4*sigma,mu+4*sigma,100)
pdf = norm.pdf(x,mu,sigma)
```

```python
plt.plot(x,pdf,color='r',label=f'N({mu},{sigma}^2)')
plt.xlabel('Exam scores')
plt.ylabel('probability density')
plt.title('Normal distributi9on')
plt.legend()
plt.grid()
```

```python
z = np.linspace(-4,4,1000)
pdf = norm.pdf(z,0,1)
```

```python
plt.plot(z,pdf,color='blue',label='standard normal distribution(mean=0,std=1)')
plt.xlabel('z-score')
plt.ylabel('probability density')
plt.title('standard normal distribution(z-distribution)')
plt.axvline(x=0,color='r',linestyle='--',label='mean(z=0)')
plt.legend()
plt.grid()
plt.show()
```

```python
np.random.seed(42)
income_data = np.random.exponential(scale=50,size=1000)
plt.figure(figsize=(8,4))
sns.histplot(income_data,bins=50,kde=True,color='r')
plt.xlabel('income($1000)')
plt.ylabel('freqency')
plt.show()
sample_means = []
n_samples = 1000
sample_size=30
for _ in range(n_samples):
    sample = np.random.choice(income_data,size=sample_size)
    sample_means.append(np.mean(sample))
sns.histplot(sample_means,bins=50,kde=True,color='b')
plt.title('sample mean')
plt.ylabel('frequency')
plt.show()
```
